## visualization

```python
sns.pairplot(clean)       
```

Top 10 countrys with most number of cases

```python
top_10_cases=clean.groupby(by='country').max([['total_cases','date']]).sort_values(by=['total_cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 countrys with most number of cases")
ax=sns.barplot(data=top_10_cases.iloc[:10],y='total_cases',x='country',linewidth=2,edgecolor='r')
```

Top 10 countrys with most number of deaths

```python
top_10_deaths=clean.groupby(by='country').max([['total_deaths','date']]).sort_values(by=['total_deaths'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 countrys with most number of cases")
ax=sns.barplot(data=top_10_deaths.iloc[:10],y='total_deaths',x='country',linewidth=2,edgecolor='b')
```

Top 10 countrys with most number of vaccinated peoples

```python
top_10_deaths=clean.groupby(by='country').max([['people_vaccinated','date']]).sort_values(by=['people_vaccinated'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
plt.title("Top 10 countrys with most number of cases")
ax=sns.barplot(data=top_10_deaths.iloc[:10],y='people_vaccinated',x='country',linewidth=2,edgecolor='g')
```
